IL( Intermediate Language)，也称CIL或者MSL，翻译成中文就是“中间语言”。IL由ECMA组织（ECMA-335标准）提供完整的定义和规范。使用 Visual Studio中的C#编译器，我们可以直接把C#源程序编译为.exe或.dll文件，但此时编译出来的程序代码并不是CPU能直接执行的二进制代码，而是IL代码。

在编译后的IL结构中，包含了 MANIFEST文件。MANIFEST是一个清单文件，主要包含程序集的一些属性，例如程序集名称、版本号、哈希算法、程序集模块，以及对外部引用程序的引用项目。
